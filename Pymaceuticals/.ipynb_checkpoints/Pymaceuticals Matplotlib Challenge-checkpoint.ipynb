```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings

# Hide warning messages in notebook
warnings.filterwarnings('ignore')

# File to Load
mouse_drug_csv = "data/mouse_drug_data.csv"
clinical_trial_csv = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse = pd.read_csv(mouse_drug_csv)

# clinical_trial_data = pd.read_csv(clinical_trial_csv)
clinical = pd.read_csv(clinical_trial_csv)

# Combine the data into a single dataset
df = pd.merge(clinical, mouse, on='Mouse ID')

# Display the data table for preview
df.head(10)
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_time_data = df[["Drug","Timepoint","Tumor Volume (mm3)"]]

drug_time_group = drug_time_data.groupby(['Drug','Timepoint'])
drug_time_series = drug_time_group["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
drug_time_df = drug_time_series.reset_index()

# Preview DataFrame
drug_time_df.head()

```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_stnd_err = drug_time_group[ ['Tumor Volume (mm3)'] ].sem()

# Convert to DataFrame
tumor_stnd_err_df = pd.DataFrame(tumor_stnd_err)
tumor_stnd_err_df.reset_index(inplace=True)
tumor_stnd_err_df.rename(columns={'Tumor Volume (mm3)':'Tumor Standard Error'}, inplace=True)

# Preview DataFrame
tumor_stnd_err_df.head()
```

```python
# Data Munging to Re-Format the Data Frames
sorted_by_tumor = drug_time_df.groupby('Drug')['Tumor Volume (mm3)'].max().reset_index()
sorted_by_tumor = sorted_by_tumor.sort_values("Tumor Volume (mm3)")
drug_time_cols_ordered = sorted_by_tumor['Drug'].unique().tolist()
tumor_pivot = drug_time_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')

# Preview that Reformatting worked
tumor_pivot

```

```python
# sort by last day tumor size to filter out similar drugs and make data chart more clear
tumor_pivot.loc[45].reset_index().sort_values(by=45)

# set relevant drugs withou noise data
relevant_drugs = ['Ramicane','Capomulin','Ceftamin','Zoniferol','Placebo','Ketapril']
```

```python
#Set Scatter Plot Legend
listLegend=relevant_drugs
markers =  ["<--","v--","s--","p--","P--","X--","D--",">--",".--","^--"]
colors =  ["#92ba18","#56ba18","#ba9718","#ba4b18","#ba3b18","#e00202"]

for i,drug in enumerate(listLegend):
    plt.errorbar(tumor_pivot.index,tumor_pivot[drug],tumor_stnd_err_df[ tumor_stnd_err_df['Drug']==drug]['Tumor Standard Error'], fmt=markers[i], markersize=12, alpha=0.6, color=colors[i], ecolor='black', elinewidth=2)

# Set Styles, Titles and Labels for our chart
plt.rcParams["figure.figsize"] = [18, 10]
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

# Create a legend for our chart
plt.legend(listLegend,loc="best")

plt.grid()
# Save the Figure
plt.savefig("Tumor-Response-to-Treatment.png")

```

## Metastatic Response to Treatment

```python
# Store the Mean metastatic Site Data Grouped by Drug and Timepoint 
metastatic_mean = pd.DataFrame(df.groupby(["Drug","Timepoint"]).mean())[['Metastatic Sites']]

# Convert to DataFrame
metastatic_mean = metastatic_mean.reset_index()

# Preview DataFrame
metastatic_mean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_sem = pd.DataFrame(df.groupby(["Drug","Timepoint"])[['Metastatic Sites']].sem())

# Convert to DataFrame
metastatic_sem = metastatic_sem.reset_index()
metastatic_sem.rename(columns={'Metastatic Sites':'Metastatic Standard Error'}, inplace=True)

# Preview DataFrame
metastatic_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frame
sorted_by_met = metastatic_mean.groupby('Drug')['Metastatic Sites'].max().reset_index()
sorted_by_met = sorted_by_met.sort_values("Metastatic Sites")

met_sites_cols_ordered = sorted_by_met['Drug'].unique().tolist()

metastatic_pivot = metastatic_mean.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
metastatic_pivot = metastatic_pivot.reindex_axis(met_sites_cols_ordered, axis=1)

# Preview that Reformatting worked
metastatic_pivot
```

```python
# Generate the Plot (with Error Bars)
metListLegend= relevant_drugs
markers =  ["<--","v--","s--","p--","P--","X--","D--",">--",".--","^--"]
colors =  ["#56ba18","#92ba18","#d8db1a","#ba4b18","#ba3b18","#e00202"]

for i,drug in enumerate(metListLegend):
    plt.errorbar(metastatic_pivot.index,metastatic_pivot[drug],metastatic_sem[ metastatic_sem['Drug']==drug]['Metastatic Standard Error'], fmt=markers[i], markersize=12, alpha=0.6, color=colors[i], ecolor='black', elinewidth=3)

# Create a legend for our chart
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastatic Sites")
plt.legend(metListLegend,loc="best")
# Show the chart
plt.grid()
# Save the Figure
plt.savefig("Metastatic-Spread-During-Treatment.png")
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint
show_columns=[
      'Drug',
      'Timepoint',
      'Mouse ID']

# Convert to DataFrame
mice_count_df = pd.DataFrame(df.groupby(["Drug","Timepoint"],as_index = False).count())[show_columns].rename(columns={'Mouse ID': 'Mouse Count'})

# Preview DataFrame
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_pivot = mice_count_df.pivot(index='Timepoint', columns='Drug', values='Mouse Count')

# Preview the Data Frame
mice_pivot
```

```python
mice_legend = relevant_drugs
markers =  ["<--","v--","s--","p--","P--","X--","D--",">--",".--","^--"]
colors =  ["#56ba18","#92ba18","#d8db1a","#ba4b18","#ba3b18","#e00202"]

for i,drug in enumerate(mice_legend):
    percentage = 100 * mice_pivot[drug] / max(mice_pivot[drug])
    plt.plot(mice_pivot.index, percentage, markers[i], color=colors[i], markersize=12, alpha=0.6, linewidth=2)

# Create a legend for our chart
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend(mice_legend,loc="best")

plt.ylim(20,103)
plt.grid()
# Save the Figure
plt.savefig("Survival During Treatment.png")
```
